####Import 

```
#import
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
df = pd.read_csv("https://raw.githubusercontent.com/ninayang2019/Machine-Learning/master/partII.csv")
df.head()
```

```
# Data Cleaning

# It seems that some of the recruiters got ZERO early_applies but eventually managed to submit candidates
# Which would leads to an infinite value of their submiiting efficiency
# I am sifting out these eight records here where early applies < submits early 
df_delete = df[df['submits_early']>df['early_applies']]
df_delete
```

# Metrics
- Metric 1 :Submitting efficency  = AVG (Number of Submits_Early of each req / Number of Early_Applicants of each req )
- Metric 2: Screening efficiency = AVG( Number of Screen Applicants of each req/ Number of Early Applicants of each req )


### Metric 1:Submitting efficiency

```
df1=df[df['submits_early']<=df['early_applies']]

# Adding a column as submitting% = number of submits_early/number of early_applies
df1['submitting%'] =df1['submits_early']/df1['early_applies']

# Fillout Nan with 0
df1.fillna(0,inplace=True)

# Calculate average submiiting efficiency of each requireter
df1.groupby('primary_recruiter')['submitting%'].mean()
```

```
#Export it as a data.frame
df1_submitting = df1[['submitting%','primary_recruiter']].groupby('primary_recruiter',as_index=True).mean()
df1_submitting.head()
# Add a rank of screening efficiency
df1_submitting['submitting_efficiency_rank'] = df1_submitting['submitting%'].rank(ascending=False)
df1_submitting = df1_submitting.reset_index()
df1_submitting
```

### Metric 2: Screening efficiency

```
df1['screening_efficiency'] = df1['screened_applies']/df1['early_applies']
df1.fillna(0,inplace=True)
df1_screening = df1[['screening_efficiency','primary_recruiter']].groupby('primary_recruiter',as_index=True).mean()
df1_screening.head()
```

```
# Export it as data.frame
df1_screening['screening_efficiency_rank'] = df1_screening['screening_efficiency'].rank(ascending=False)
df1_screening = df1_screening.reset_index()
df1_screening
```

# Recruiter ranking


```
# Join two data.frames to have two types of rank together
df_rank = df1_screening.merge(df1_submitting,how='inner')
df_rank

# Add a compounded rank = rank of submitting efficiency +rank of screening efficiency rank
# The smaller the rank number, the higher efficiency of both screening and submitting
df_rank['compounded_rank'] = df_rank['screening_efficiency_rank']+ df_rank['submitting_efficiency_rank']
df_rank = df_rank.sort_values('compounded_rank').reset_index()
```
