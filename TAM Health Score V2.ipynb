```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_target = pd.read_csv('TARGET.csv')
df1=pd.read_csv('m1.csv')
df2=pd.read_csv('m2.csv')
df3=pd.read_csv('m-zendesk.csv')
df4=pd.read_csv('paids.csv')
df5=pd.read_csv('adoption-utf.csv')
df6=pd.read_csv('m6.csv')
df4
```

```python
df12=pd.merge(df1,df2,how='outer',on=['sfdc_acct_name'])
df34=pd.merge(df3,df4,how='outer',on=['sfdc_acct_name'])
df1234=pd.merge(df12,df34,how='outer',on=['sfdc_acct_name'])
df56=pd.merge(df5,df6,how='outer',on=['sfdc_acct_name'])
df56t=pd.merge(df56,df_target,how='outer',on=['sfdc_acct_name'])
mdf=pd.merge(df1234,df56t,how='outer',on=['sfdc_acct_name'])
print(mdf)
```

```python
corr=mdf.corr()
fig=corr.style.background_gradient(cmap='coolwarm')
fig
```

```python
df = pd.DataFrame(np.random.random((6, 14)), columns=list(mdf.columns.tolist()))
fig, ax = plt.subplots(figsize=(16, 12))
sns.heatmap(df.corr(method='pearson'), annot=True, fmt='.4f', 
            cmap=plt.get_cmap('coolwarm'), cbar=False, ax=ax)
ax.set_yticklabels(ax.get_yticklabels(), rotation="horizontal")
plt.savefig('result.png', bbox_inches='tight', pad_inches=0.0)
```
